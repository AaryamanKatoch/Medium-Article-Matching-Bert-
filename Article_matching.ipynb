```python
df = pd.read_csv('articles.csv')
```

```python
df.head()
```

```python
articles = pd.DataFrame(df,columns=['text'])
```

```python
#removing punctuation and special characters
articles["text"] = articles["text"].str.replace('\n',' ',regex=False)
articles["text"] = articles["text"].str.replace('-', '',regex=False)
articles["text"] = articles["text"].str.replace('_', '',regex=False)
articles["text"] = articles["text"].str.replace('(', '',regex=False)
articles["text"] = articles["text"].str.replace(')', '',regex=False)
articles["text"] = articles["text"].str.replace('@', '',regex=False)
articles["text"] = articles["text"].str.replace('#', '',regex=False)
articles["text"] = articles["text"].str.replace('$', '',regex=False)
articles["text"] = articles["text"].str.replace('%', '',regex=False)
articles["text"] = articles["text"].str.replace('^', '',regex=False)
articles["text"] = articles["text"].str.replace('&', '',regex=False)
articles["text"] = articles["text"].str.replace('*', '',regex=False)
articles["text"] = articles["text"].str.replace('+', '',regex=False)
articles["text"] = articles["text"].str.replace('=', '',regex=False)
articles["text"] = articles["text"].str.replace('?', '',regex=False)
articles["text"] = articles["text"].str.replace('|', '',regex=False)
articles["text"] = articles["text"].str.replace('\\', '',regex=False)
articles["text"] = articles["text"].str.replace('/', '',regex=False)
articles["text"] = articles["text"].str.replace(':', '',regex=False)
articles["text"] = articles["text"].str.replace(':', '',regex=False)
articles["text"] = articles["text"].str.replace('<', '',regex=False)
articles["text"] = articles["text"].str.replace('>', '',regex=False)
articles["text"] = articles["text"].str.replace('[', '',regex=False)
articles["text"] = articles["text"].str.replace(']', '',regex=False)
articles["text"] = articles["text"].str.replace('{', '',regex=False)
articles["text"] = articles["text"].str.replace('}', '',regex=False)
```

```python
#removing stop words
stop = stopwords.words('english')
articles['articles_final'] = articles["text"].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop)]))
```

```python
# using BERT to create word embeddings
document_embeddings = bert_model.encode(articles['articles_final'])
```

```python
#view shape
document_embeddings.shape
```

```python
#get embeddings
document_embeddings[0]
```

```python
#similarity function for cosine,Euclidean distance and Polynomial
def get_similar(input_str,sim_type,num):
    str_embeddings = bert_model.encode(input_str)
    res=np.array([])
    if sim_type=='Cosine':
        matrix=cosine_similarity(document_embeddings,str_embeddings.reshape(1, -1))
        res=np.argsort(matrix,axis=None)[::-1][:num:]
    elif sim_type=='Euclidean':
        matrix=euclidean_distances(document_embeddings,str_embeddings.reshape(1, -1))
        res=np.argsort(matrix,axis=None)[:num:]
    elif sim_type=='Polynomial':
        matrix=polynomial_kernel(document_embeddings,str_embeddings.reshape(1, -1))
        res=np.argsort(matrix,axis=None)[::-1][:num:]

#generating top n ranked documents -  n given by the user        
    for i in res:
        print("\n Similarity Value: ",matrix[i],'\n Article: ')
        print(articles["articles_final"].iloc[i])
        print(i)
    
```
